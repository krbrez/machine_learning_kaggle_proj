```python
# Various Imports
import numpy as np
import matplotlib .pyplot as plt
import pandas as pd
import cv2
import os
from tqdm import tqdm
```

```python
# Get the data values from the csv
df = pd.read_csv("train.csv") 
```

```python
# Convert the pandas datafram to a numpy array
nparry = df.to_numpy()
# There are 25,361 images and corresponding labels(thus an array of 25361 by 2)
print(nparry.shape)
```

```python
# Seperate the whale image IDs and the labels(y) into two seperate arrays
whale_image_IDs = nparry[:,0]
labels = nparry[:,1]
```

```python
# One hot encode the lables ( should be in the form [1, 0, 0 ... 0, 0, 0] where the 1 represents the correct value)
yhat = np.zeros((labels.shape[0], np.unique(labels).shape[0])) # For each y label, create an array of zeros 5005 long

indexing_array = {}
indexing_array_reveresd = {}
index = 0
true_index = 0

for x in tqdm(labels):
    arr = np.zeros(np.unique(labels).shape[0]) #Create an array of zeros 5005 long
    # If we already know the index of that label, then that is what we want, otherwise add it to the list with a new index
    if x in indexing_array:
        target_index = indexing_array.get(x) # Just retrieve the correct index value
    else:
        indexing_array.update({x: index}) # Add the index to the list of indexes and increment
        target_index = index
        index += 1
    true_index += 1
    
    arr[target_index] = 1
    yhat[true_index - 1] = arr
```

```python
# At this point yhat should be a 25361 x 5005 matrix that is one hot encoded
print(yhat.shape)
print(indexing_array.get('000a6daec.jpg'))
print(np.where(yhat[8] == 1))
```

```python
# Declare which folder has the training images
DATADIR = "./humpback-whale-identification"
category = "train"
path = os.path.join(DATADIR, category)
```

```python
X_grey_100x100 = np.empty((25361, 100, 100))
print(X_grey_100x100.shape)
```

```python
# This dataset will be black and white, and of size 100x100

index = 0
for img in tqdm(os.listdir(path)):
    img_array = cv2.imread(os.path.join(path,img), cv2.IMREAD_GRAYSCALE)
    new_array = cv2.resize(img_array, (100, 100))
    row_vec = new_array[np.newaxis, :]
    X_grey_100x100[index] = row_vec
    #plt.imshow(new_array)
    #plt.show()
    index += 1
print(X_grey_100x100[0].shape)
print(X_grey_100x100.shape)
X_grey_100x100 = np.empty((25361, 100, 100))
print(X_grey_100x100.shape)np.save('X_grey_100x100', X_grey_100x100)
```

```python
# Because we do not have a labeled testing set, we will divide our data as such (Lets not actually run this part)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_color_32x32, yhat, test_size=0.5, random_state=42)
```

```python
X_train = X_color_32x32
y_train = yhat
```

```python
# Print the new sizes of our testing and training sets
print('Size of X_train -> ', X_train.shape)
print('Size of y_train -> ', y_train.shape)
#print('Size of X_test -> ', X_test.shape )
#print('Size of y_test -> ', y_test.shape)
```

```python
# Normalize the pixel values to be between 0 and 1
X_train = X_train / 255
X_test = X_test / 255
```

```python
#Build the CNN
from keras.models import Sequential
from keras.layers import Dense, Flatten, Conv2D, MaxPooling2D

#Create the architecture
model = Sequential()

#Convolution layer
model.add( Conv2D(32, (5, 5), activation='relu', input_shape=(32,32,3)))

#MaxPooling layer
model.add( MaxPooling2D(pool_size=(2,2)))

#Convolution layer
model.add( Conv2D(32, (5, 5), activation='relu'))

#MaxPooling layer
model.add( MaxPooling2D(pool_size=(2,2)))

#Flatten layer
model.add(Flatten())

model.add(Dense(5000, activation='relu'))
model.add(Dense(5005, activation='softmax'))
```

```python
# Compile the model
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Train the model, we can actually tell keras to split the data into a 30% validation set, which is cool
hist = model.fit(X_train, y_train, batch_size=256, epochs=10, validation_split=0.3)
```

```python
#Get the models accuracy
print(model.evaluate(X_test, y_test)[1])
```

```python
#Visualize the models accuracy
plt.plot(hist.history['accuracy'])
plt.plot(hist.history['val_accuracy'])
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Val'], loc='upper left')
plt.show()
```

```python
#Visualize the models accuracy
plt.plot(hist.history['loss'])
plt.plot(hist.history['val_loss'])
plt.title('Model Loss')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Val'], loc='upper right')
plt.show()
```
